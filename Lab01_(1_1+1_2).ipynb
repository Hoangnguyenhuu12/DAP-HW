```python
import pandas as pd
df = pd.read_csv('/titanic.csv')
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df['embarked'] = df['embarked'].replace({'C': 'Cherbourg', 'Q': 'Queenstown', 'S': 'Southampton'}).astype('category')

df['sex'] = df['sex'].replace({1: 'Female', 0: 'Male'}).astype('category')

df['survived'] = df['survived'].replace({1: 'Yes', 0: 'No'}).astype('category')

df['pclass'] = df['pclass'].replace({1: '1st', 2: '2nd', 3: '3rd'}).astype('category')
df['pclass'] = pd.Categorical(df['pclass'], ['1st', '2nd', '3rd'], ordered=True)

df.head()
```
